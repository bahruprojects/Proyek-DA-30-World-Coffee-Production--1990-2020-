```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1. Membaca dataset
df = pd.read_csv(r"C:\Users\OPTION\Downloads\Python DA\[BELUM] Projects 30_World Coffee Production (1990 - 2020) By Alfin Bahru Rahmika Umar\Coffee_production.csv")
df
```

```python
# Menampilkan informasi dasar
print("Dimensi awal dataset:", df.shape)
```

```python
df.info()
df.describe()
```

```python
# Tampilkan informasi awal
print("Shape sebelum cleaning:", df.shape)
print("Jumlah duplikat:", df.duplicated().sum())
print("Missing values per kolom:\n", df.isnull().sum())

# Deteksi outlier menggunakan IQR untuk kolom numerik
numeric_cols = df.select_dtypes(include=[np.number]).columns

def detect_outliers(data, col):
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return data[(data[col] < lower_bound) | (data[col] > upper_bound)]

outlier_counts = {}
for col in numeric_cols:
    outliers = detect_outliers(df, col)
    outlier_counts[col] = outliers.shape[0]
print("Jumlah outlier per kolom:", outlier_counts)
```

```python
# Melakukan Data Cleaning:
# 1. Hapus duplikat dan missing values
df_cleaned = df.drop_duplicates().dropna()

# 2. Hapus outlier untuk setiap kolom numerik
for col in numeric_cols:
    Q1 = df_cleaned[col].quantile(0.25)
    Q3 = df_cleaned[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df_cleaned = df_cleaned[(df_cleaned[col] >= lower_bound) & (df_cleaned[col] <= upper_bound)]

print("Shape setelah cleaning:", df_cleaned.shape)
```

```python
# Visualisasi: Histogram dan Box Plot untuk masing-masing kolom numerik
for col in numeric_cols:
    plt.figure(figsize=(12,5))
    
    plt.subplot(1,2,1)
    sns.histplot(df_cleaned[col], kde=True, bins=15)
    plt.title(f'Histogram of {col}')
    
    plt.subplot(1,2,2)
    sns.boxplot(x=df_cleaned[col])
    plt.title(f'Boxplot of {col}')
    plt.tight_layout()
    plt.show()
```

```python
# Perbandingan statistik deskriptif sebelum dan sesudah cleaning
print("Statistik Deskriptif Sebelum Cleaning:")
print(df.describe())

print("Statistik Deskriptif Setelah Cleaning:")
print(df_cleaned.describe())
```

```python
df_cleaned.to_csv('Coffee_production-CLEANED.csv', index=False)
```

```python
# Total Global Coffee Production
total_global_production = df_cleaned['Total_production'].sum()
print("Total Global Coffee Production in 30 Years:", total_global_production)
```

```python
# Bar Chart: Total Coffee Production Per Country
# Asumsi setiap negara hanya muncul satu kali; jika tidak, lakukan grouping
bar_data = df_cleaned.groupby('Country')['Total_production'].sum().reset_index()

plt.figure(figsize=(12,6))
sns.barplot(data=bar_data, x='Country', y='Total_production', palette='viridis')
plt.title("Total Coffee Production Per Country")
plt.xlabel("Country")
plt.ylabel("Total Production")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# Pie Chart: Percentages of Coffee Type
coffee_type_data = df_cleaned['Coffee type'].value_counts()
plt.figure(figsize=(8,8))
plt.pie(coffee_type_data, labels=coffee_type_data.index, autopct='%1.1f%%', startangle=140)
plt.title("Percentages of Coffee Type")
plt.show()
```

```python
# Line Chart: Coffee Production Per Year
# Identifikasi kolom tahun (mengandung '/')
year_columns = [col for col in df_cleaned.columns if '/' in col and col not in ['Coffee type']]
yearly_production = df_cleaned[year_columns].sum()

plt.figure(figsize=(14,7))
sns.lineplot(x=yearly_production.index, y=yearly_production.values, marker='o')
plt.title("Coffee Production Per Year")
plt.xlabel("Year")
plt.ylabel("Production")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
